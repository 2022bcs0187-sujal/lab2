<h2><b>Linear Regression Implementation</b></h2>

```python
from sklearn.metrics import mean_squared_error, r2_score

def run_experiment(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    return mse, r2
```

<h4><b>EXP-01: Linear Regression (All Features, No Scaling)</b></h4>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = df.drop(columns=['quality'])
y = df['quality']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
mse1, r2_1 = run_experiment(model, X_train, X_test, y_train, y_test)

mse1, r2_1
```

<h4><b>EXP-02: Ridge Regression + Standardization + Feature Selection</h4></b>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

corr = df.corr()['quality'].abs()
selected_features = corr[corr > 0.2].index.drop('quality')

X_sel = df[selected_features]

X_train, X_test, y_train, y_test = train_test_split(
    X_sel, y, test_size=0.3, random_state=42
)

model = Ridge(alpha=1.0)
mse2, r2_2 = run_experiment(model, X_train, X_test, y_train, y_test)

mse2, r2_2
```

<h4><b>EXP-03: Random Forest (50 Trees)</h4></b>

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

mse3, r2_3 = run_experiment(model, X_train, X_test, y_train, y_test)

mse3, r2_3
```

<h4><b>Random Forest (100 Trees, Feature Subset)</h4></b>

```python
model = RandomForestRegressor(n_estimators=100, max_depth=15, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(
    X_sel, y, test_size=0.2, random_state=42
)

mse4, r2_4 = run_experiment(model, X_train, X_test, y_train, y_test)

mse4, r2_4
```

## Task 5: Analysis and Interpretation

### 1. Which experiment produced the lowest MSE?
**Experiment EXP-03** produced the lowest Mean Squared Error (MSE) with a value of **0.3269**.  
This indicates that the Random Forest model with 50 trees and depth 10 achieved the most accurate predictions among all experiments.

---

### 2. Which experiment achieved the highest R² score?
**Experiment EXP-03** achieved the highest R² score of **0.4996**, meaning it explained nearly 50% of the variance in the wine quality scores.

---

### 3. How did pre-processing affect model performance?
Pre-processing through standardization and regularization (EXP-02) resulted in **lower performance** compared to the baseline linear regression model (EXP-01).  
This suggests that scaling and regularization did not significantly benefit the linear model for this dataset, possibly due to already well-scaled features or loss of important variance.

---

### 4. Did feature selection improve or degrade performance?
Feature selection had a **mixed impact**:
- It **degraded performance** for the linear model (EXP-02), reducing both MSE and R².
- It **slightly improved performance** for the Random Forest model (EXP-04) compared to linear models, but did not surpass EXP-03 which used all features.

---

### 5. Identify the best overall model with justification.
The **Random Forest model in EXP-03** is the best overall model as it achieved the **lowest MSE** and the **highest R² score**.  
Its superior performance is attributed to its ability to capture complex, non-linear relationships in the data without requiring extensive preprocessing.


<h1><b>Why is structured experiment tracking critical for scalable and reproducible machine
learning workflows?</b><h1>

Structured experiment tracking is critical for scalable and reproducible machine learning workflows because it ensures that every model’s configuration, data split, and evaluation metric is clearly documented. As seen in this lab, multiple experiments with different models, preprocessing steps, and hyperparameters can quickly become difficult to manage without proper tracking. Systematic tracking enables fair comparison between experiments and helps identify the best-performing model reliably. It also allows experiments to be reproduced or extended in the future without ambiguity. As projects scale and involve more experiments or collaborators, structured tracking becomes essential to avoid errors and improve collaboration. Ultimately, it motivates the adoption of MLOps tools for automation and reliability.
